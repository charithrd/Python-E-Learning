```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
```

> ## <font color='lightblue'>Importing glob module to assist in finding file paths matching a specified pattern</font>

- glob function within glob module is used to search for '/content/K+N Seafreight WK*.xlsm' pattern within the content directory and return all the matching files.

```python
import glob
glob.glob('/content/K+N Seafreight WK*.xlsm')
```

> ### <font color='lightblue'>Initialising an Empty List and Append Files</font>
- An empty list is created using 'comb_dfs = []'
- For loop iterates over each file path that matches the glob function pattern and assign each file name to the variable 'one_filename'.
- Pandas read_excel function is used to read the files in 'one_filename' and load into a dataframe. This dataframe is then assigned to the 'new_df' variable.
- Finally, 'new_df' dataframe is added to the 'comb_dfs' list.

```python
comb_dfs = []
for one_filename in glob.glob('/content/K+N Seafreight WK*.xlsm'):
    print(f'loading {one_filename}')
    new_df = pd.read_excel(one_filename)
    comb_dfs.append(new_df)
```

```python
# Pandas concatenate function is used to combine multiple dataframes within 'comb_dfs'
df =pd.concat(comb_dfs)
```

> ### <font color='lightblue'>List comprehension using LLM's</font>

- Code that we used to Initialise an empty list and appendthe files to a dataframe is optimised using List Comprehension technique.
- To achieve this objective we have seeked recomendations from LLM.
- Similar output is achieved with minimal coding.

```python
df = pd.concat([pd.read_excel(one_filename)
      for one_filename in glob.glob('/content/K+N Seafreight WK*.xlsm')
      if not print(f'loading {one_filename}')
])
```

```python
# Dimensions of the dataframe using shape function.
df.shape
```

```python
# First 15 rows of the dataframe is retrieved using head() function.
df.head(15)
```

### Observations
- With the head function output we can observe that the dataset headers exist in row 1 (index 0)
- Hence, we are using iloc function to target row 1 'iloc[0]' as headers and subsequently drop row 1 since it's the new header.

```python
# Set the first row as the header.
df.columns = df.iloc[0]

# Drop the first row since it's now the header.
df_trim = df.drop(0).reset_index(drop=True)

# Display the DataFrame to confirm the changes.
df_trim.head(15)
```

```python
# Returning the data types of each column in 'df_trim' DataFrame.
df_trim.dtypes
```

### Observations
- At a glance we can spot that data in each column is created as an object.
- With the head() function we have noticed that columns like 'Ocean Freight USD' are numerical columns.
- Hence, the numerical columns within this dataset need converting to data type 'int' or 'floats' where necessary.


```python
# Before converting numerical columns, we first check for any extra spaces or inconsistencies in the column names.
print(df_trim.columns)
```

- It is visible that certain columns such as ' ENS Fee ' has leading and trailing whitespaces and some columns such as ' Ocean Freight USD' has only leading whitespace.

- To eliminate the inconsistency and clean the whitespace we are using strip() function.

```python
# Remove any leading/trailing whitespace from column names and printing the refined column names.
df_trim.columns = df_trim.columns.str.strip()
print(df_trim.columns)
```

```python
# Converting the columns datatype as necessary, while replacing null values with zero.
df_trim['Ocean Freight USD'] = pd.to_numeric(df_trim['Ocean Freight USD'], errors='coerce').fillna(0).astype(float)
df_trim['Hazardous'] = pd.to_numeric(df_trim['Hazardous'], errors='coerce').fillna(0).astype(float)
df_trim['Reefer'] = pd.to_numeric(df_trim['Reefer'], errors='coerce').fillna(0).astype(float)
df_trim['ISPS USD'] = pd.to_numeric(df_trim['ISPS USD'], errors='coerce').fillna(0).astype(float)
df_trim['ENS Fee'] = pd.to_numeric(df_trim['ENS Fee'], errors='coerce').fillna(0).astype(float)
df_trim['Sulphur Chg'] = pd.to_numeric(df_trim['Sulphur Chg'], errors='coerce').fillna(0).astype(float)
df_trim['GOH Surcharge'] = pd.to_numeric(df_trim['GOH Surcharge'], errors='coerce').fillna(0).astype(float)
df_trim['USD Total'] = pd.to_numeric(df_trim['USD Total'], errors='coerce').fillna(0).astype(float)
df_trim['No of Drops'] = pd.to_numeric(df_trim['No of Drops'], errors='coerce').fillna(0).astype(int)
df_trim['Terminal Handling GBP'] = pd.to_numeric(df_trim['Terminal Handling GBP'], errors='coerce').fillna(0).astype(float)
df_trim['Port Security GBP'] = pd.to_numeric(df_trim['Port Security GBP'], errors='coerce').fillna(0).astype(float)
df_trim['Documents GBP'] = pd.to_numeric(df_trim['Documents GBP'], errors='coerce').fillna(0).astype(float)
df_trim['Customs Clearance GBP'] = pd.to_numeric(df_trim['Customs Clearance GBP'], errors='coerce').fillna(0).astype(float)
df_trim['Haulage GBP'] = pd.to_numeric(df_trim['Haulage GBP'], errors='coerce').fillna(0).astype(float)
df_trim['DRS'] = pd.to_numeric(df_trim['DRS'], errors='coerce').fillna(0).astype(float)
df_trim['Port Inf'] = pd.to_numeric(df_trim['Port Inf'], errors='coerce').fillna(0).astype(float)
df_trim['Energy Adjustment and Fuel Duty Recovery GBP'] = pd.to_numeric(df_trim['Energy Adjustment and Fuel Duty Recovery GBP'], errors='coerce').fillna(0).astype(float)
df_trim['Gen Set'] = pd.to_numeric(df_trim['Gen Set'], errors='coerce').fillna(0).astype(float)
df_trim['HAZ Surcharge'] = pd.to_numeric(df_trim['HAZ Surcharge'], errors='coerce').fillna(0).astype(float)
df_trim['Rework'] = pd.to_numeric(df_trim['Rework'], errors='coerce').fillna(0).astype(float)
df_trim['GBP Total'] = pd.to_numeric(df_trim['GBP Total'], errors='coerce').fillna(0).astype(float)
df_trim['Gate In (Freight)'] = pd.to_numeric(df_trim['Gate In (Freight)'], errors='coerce').fillna(0).astype(int)
df_trim['Delivery Month'] = pd.to_numeric(df_trim['Delivery Month'], errors='coerce').fillna(0).astype(int)
```

```python
# Dropping not required columns from the dataframe.
df_trim.drop(['File Reference','FAK Y/N?'],axis=1,inplace=True)
```

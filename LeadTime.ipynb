> ## <font color='lightblue'>Importing glob module to assist in finding file paths matching a specified pattern</font>

- glob function within glob module is used to search for '/content/K+N Seafreight WK*.xlsm' pattern within the content directory and return all the matching files.

```python
import glob
glob.glob('/content/K+N Seafreight WK*.xlsm')
```

> ### <font color='lightblue'>Initialising an Empty List and Append Files</font>
- An empty list is created using 'comb_dfs = []'
- For loop iterates over each file path that matches the glob function pattern and assign each file name to the variable 'one_filename'.
- Pandas read_excel function is used to read the files in 'one_filename' and load into a dataframe. This dataframe is then assigned to the 'new_df' variable.
- Finally, 'new_df' dataframe is added to the 'comb_dfs' list.

```python
comb_dfs = []
for one_filename in glob.glob('/content/K+N Seafreight WK*.xlsm'):
    print(f'loading {one_filename}')
    new_df = pd.read_excel(one_filename)
    comb_dfs.append(new_df)
```

```python
# Pandas concatenate function is used to combine multiple dataframes within 'comb_dfs'
df =pd.concat(comb_dfs)
```

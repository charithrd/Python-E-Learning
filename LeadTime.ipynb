> ## <font color='lightblue'>Importing glob module to assist in finding file paths matching a specified pattern</font>

- glob function within glob module is used to search for '/content/K+N Seafreight WK*.xlsm' pattern within the content directory and return all the matching files.

```python
import glob
glob.glob('/content/K+N Seafreight WK*.xlsm')
```

> ### <font color='lightblue'>Initialising an Empty List and Append Files</font>
- An empty list is created using 'comb_dfs = []'
- For loop iterates over each file path that matches the glob function pattern and assign each file name to the variable 'one_filename'.
- Pandas read_excel function is used to read the files in 'one_filename' and load into a dataframe. This dataframe is then assigned to the 'new_df' variable.
- Finally, 'new_df' dataframe is added to the 'comb_dfs' list.

```python
comb_dfs = []
for one_filename in glob.glob('/content/K+N Seafreight WK*.xlsm'):
    print(f'loading {one_filename}')
    new_df = pd.read_excel(one_filename)
    comb_dfs.append(new_df)
```

```python
# Pandas concatenate function is used to combine multiple dataframes within 'comb_dfs'
df =pd.concat(comb_dfs)
```

> ### <font color='lightblue'>List comprehension using LLM's</font>

- Code that we used to Initialise an empty list and appendthe files to a dataframe is optimised using List Comprehension technique.
- To achieve this objective we have seeked recomendations from LLM.
- Similar output is achieved with minimal coding.

```python
df = pd.concat([pd.read_excel(one_filename)
      for one_filename in glob.glob('/content/K+N Seafreight WK*.xlsm')
      if not print(f'loading {one_filename}')
])
```

```python
# Dimensions of the dataframe using shape function.
df.shape
```

```python
# First 15 rows of the dataframe is retrieved using head() function.
df.head(15)
```

### Observations
- With the head function output we can observe that the dataset headers exist in row 1 (index 0)
- Hence, we are using iloc function to target row 1 'iloc[0]' as headers and subsequently drop row 1 since it's the new header.

```python
# Set the first row as the header.
df.columns = df.iloc[0]

# Drop the first row since it's now the header.
df_trim = df.drop(0).reset_index(drop=True)

# Display the DataFrame to confirm the changes.
df_trim.head(15)
```

```python
# Returning the data types of each column in 'df_trim' DataFrame.
df_trim.dtypes
```

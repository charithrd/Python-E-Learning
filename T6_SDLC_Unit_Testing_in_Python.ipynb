# **1. Unit Testing in Python**
[Return to contents](#contents)


**Unit testing** is an essential part of software development that involves testing individual components of your software, commonly referred to as "units," to ensure they work as intended. This practice allows you to catch bugs early, simplify debugging, and improve the design of your code.

**Objectives**
By the end of this lesson, you should be able to:

1. Understand the purpose of unit testing
2. Write simple manual tests in Python
3. Use the unittest framework in Python
4. Grasp core concepts like Test Fixtures, Test Suites, and Mocking

### **1.1 Why Do We Need Unit Testing?**
**The Purpose**

1. **Validate Code:** Ensure that your code works as expected.
2. **Prevent Breakdowns:** Make sure that new changes don't break existing functionality.
3. **Simplify Debugging:** Easier to identify which piece of code is failing.
4. **Improve Design**: Forces you to write modular and clean code.

**Real-World Analogy**

Imagine you are building a car. Wouldn't you test each part like the brakes, the engine, and the lighting system individually before assembling the full car? Unit testing serves a similar purpose in software engineering.

# **3. Using the 'unittest' library in Python**

Pythonâ€™s unittest library makes it easier to write systematic tests.

**Basic Example**

```python
import unittest

def add(a, b):
    return a + b

class TestAddition(unittest.TestCase):

    def test_add_positive_numbers(self):
        self.assertEqual(add(3, 5), 8)

    def test_add_negative_numbers(self):
        self.assertEqual(add(-1, -1), -2)
```

**What did we just do?**


The code imports Python's `unittest` module to create unit tests.

It defines a function `add` that returns the sum of two numbers. The class `TestAddition` contains two tests for this function: one to check the sum of two positive numbers and another for two negative numbers. These tests ensure the `add` function works correctly for the given scenarios.

To run the tests, we could:
1. **Option 1** - save the code in a *.py file and use the command `python -m unittest <filename>` from the command line or terminal.
2. **Option 2** - In a separate code block, run `unittest.main(argv=[''], verbosity=2, exit=False)`

### **3.1 Now let's run our Unit Tests**

```python
unittest.main(argv=[''], verbosity=2, exit=False)
```

### **What did we just do?**

This line runs the unit tests within the script, with a specified verbosity level of 2 (detailed output) and prevents the program from terminating after running the tests.

### **Understanding the Output**

- **Both tests**, **test_add_negative_numbers** and **test_add_positive_numbers**, within the TestAddition class passed successfully, as indicated by the ... ok.
- **Ran 2 tests in 0.011s** provides a summary, indicating that 2 tests were run, and they took a total of 0.011 seconds to complete.
- The **OK** indicates that all tests passed successfully. The last line (<unittest.main.TestProgram at 0x79c41244b4c0>) is a memory reference to the test program instance. It's displayed because exit=False in the unittest.main() call.

### **Let's look at another example**

**Example: Testing a Calculator Class**

Let's take an example where we test a simple Calculator class:

```python
# calculator.py
class Calculator:
    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b
```

#### **Testing the Calculator class**

```python
# test_calculator.py
import unittest

class TestCalculator(unittest.TestCase):

    def test_add(self):
        calc = Calculator()
        self.assertEqual(calc.add(2, 3), 5)

    def test_subtract(self):
        calc = Calculator()
        self.assertEqual(calc.subtract(5, 3), 2)
```

```python
unittest.main(argv=[''], verbosity=2, exit=False)
```

### **3.2 Types of Assertions**

The unittest framework provides a variety of assertion methods to check for various conditions.

- **assertEqual(a, b**): Check a == b
- **assertNotEqual(a, b)**: Check a != b
- **assertTrue(x)**: Check that x is True
- **assertFalse(x)**: Check that x is False

And many more. You'll choose the assertion method that best fits the scenario you are testing.

### **Test Fixtures**
Test Fixtures are basically setup routines which help in laying down a foundation for running tests. For example, creating temporary databases.

Fixtures are used for any setup code that needs to run before a test. For example, you might need to connect to a database or create some objects in memory.

- setUp: Run before each test method.
- tearDown: Run after each test method.
- setUpClass: Run once before any test methods in a class.
- tearDownClass: Run once after all test methods in a class.

### **Mocking**
In unit testing, you may need to mock the behaviour of external modules or costly resources.

Mocking is essential when you want to isolate the unit of work and simulate behaviour of external dependencies.

```python
from unittest.mock import Mock

# Create a mock object
json = Mock()
json.loads.return_value = {'mock_key': 'mock_value'}

# Use the mock object
data = json.loads('some data')
print(data)  # Output: {'mock_key': 'mock_value'}
```

## **4.1 Summary - Why Test Driven Development (TDD)?**
It's a methodology that encourages writing tests before the actual implementation.

It's a cycle of:
- Write a test that fails
- Write minimal code to pass the test
- Refactor the code
